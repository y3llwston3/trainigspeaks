# Appling linear regression

```python
# aplying linear regression

def loss_function(m, b, points):
    total_error = 0 
    for i in range(len(points)):
        x = points.iloc[i].PowerAverage
        y = points.iloc[i].HeartRateAverage
        total_error += (y - (m * x + b))**2
        return total_error / float(len(points))
    
def gradient_descent(m_now , b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].PowerAverage
        y = points.iloc[i].HeartRateAverage
        
        m_gradient +=(2/n) * x * (y -(m_now * x + b_now))
        b_gradient +=(2/n) * (y -(m_now * x + b_now))
        
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    
    return m, b

m = 0
b = 0
L = 100
epochs = 1000

for i in range (epochs):
    if i % 50 == 0:
        print(f"Epoch:{i}")
    m, b = gradient_descent(m, b, files, L)
    
print(m, b)

plt.scatter(files.PowerAverage, files.HeartRateAverage, color="black")
plt.plot(list(range(150, 270)), [m * x + b for x in range(150, 270)], color ="green")

plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
x = files.PowerAverage.values.reshape(-1, 1)
y = files.HeartRateAverage.values

# Fit linear regression model
model = LinearRegression().fit(x, y)

# Make predictions
y_pred = model.predict(x)

# Plot scatter plot and linear regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title('Power vs Heart Rate')
plt.xlabel('Power Average')
plt.ylabel('Heart Rate Average')
plt.grid(True)
plt.show()
```
